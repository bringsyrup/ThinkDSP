```python
import matplotlib.pyplot as plt
import numpy as np
import thinkdsp as dsp

def synthesize2(amps, freqs, ts):
    args = np.outer(ts, freqs)
    M = np.exp(np.pi * 1.4j * args)
    ys = np.dot(M, amps)
    return ys

amps = np.array([0.6, 0.25, 0.1, 0.05]) 
freqs = np.array([100, 200, 300, 400])
ts = np.linspace(0, 1, 11025)

amps2 = amps * np.exp(2j)

ys_noOffset = synthesize2(amps, freqs, ts)
ys_offset = synthesize2(amps2, freqs, ts)
n = 11025/15
plt.plot(ts[:n], ys_noOffset[:n])
plt.plot(ts[:n], ys_offset[:n], 'c')
plt.legend(["phi = 0", "phi = 2"])
plt.show()
```

You can see that offsetting the phase did shift the original wave by the phase offset, but it did so differently to each frequency component, which makes the wave look different. If we want to have a uniform change, we need to factor in the different frequencies by multiplying them by the phase offset.

```python
amps3 = amps * np.exp(2j * freqs)

ys_offset2 = synthesize2(amps3, freqs, ts)
n = 11025/15
plt.plot(ts[:n], ys_noOffset[:n])
plt.plot(ts[:n], ys_offset2[:n], 'c')
plt.legend(["phi = 0", "phi = 2"])
plt.show()
```

2) implement dft from wikipedia

I'm defining all of the dft implementations so I can test them...

```python
def DFT(y):
    N = len(y)
    dft = [complex(0)] * N
    for k in range(N):
        s = complex(0) 
        for t in range(N):
            s += y[t] * np.exp(-2j * np.pi * t * k / N)
        dft[k] = s
    return dft

def dspDFT(y):
    N = len(y)
    ts = np.arange(N) / N
    freqs = np.arange(N)
    args = np.outer(ts, freqs)
    M = np.exp(2j * np.pi * args)
    amps = M.conj().transpose().dot(y)
    return amps

def npdft(y):
    return np.fft.fft(y)
```

Now I'm gonna test all three...

```python
from scipy.stats import linregress

signal = dsp.UncorrelatedGaussianNoise()
max_n = 12

"""test my summation dft"""
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, max_n)
results = []
for N in ns:
    ys1 = noise.ys[:N]
    result = %timeit -o DFT(ys1)
    results.append(result)
bests = [result.best for result in results]
x = np.log(ns)
y = np.log(bests)
t = linregress(x,y)
ns1 = ns
bests1 = bests
slope1 = t[0]
intercept1 = t[1]
r1 = t[2]
```

```python
"""test thinkdsp's dft"""
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, max_n)
results = []
for N in ns:
    ys2 = noise.ys[:N]
    result = %timeit -o dspDFT(ys2)
    results.append(result)
bests = [result.best for result in results]
x = np.log(ns)
y = np.log(bests)
t = linregress(x,y)
ns2 = ns
bests2 = bests
slope2 = t[0]
intercept2 = t[1]
r2 = t[2]

"""test numpy's fft"""
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, max_n)
results = []
for N in ns:
    ys3 = noise.ys[:N]
    result = %timeit -o npdft(ys3)
    results.append(result)
bests = [result.best for result in results]
x = np.log(ns)
y = np.log(bests)
t = linregress(x,y)
ns3 = ns
bests3 = bests
slope3 = t[0]
intercept3 = t[1]
r3 = t[2]

plt.loglog(ns1, bests1, "-b")
plt.loglog(ns2, bests2, "-r")
plt.loglog(ns3, bests3, "-k")
plt.xlabel("number of samples")
plt.ylabel("best execution times")
plt.legend(["DFT", "dspDFT", "npdft"])
plt.show()
```

3) cooley-tukey fft algorithm

```python
def FFT(y):
    N = len(y)
    if N > 1 and N % 2 > 0:
        raise ValueError("gimme a power of 2!")
    if N <= 1: 
        return y
    else:
        even = FFT(y[0::2])
        odd =  FFT(y[1::2])
        T = [np.exp(-2j * np.pi * k/N) * odd[k] for k in xrange(N/2)]
        return [even[k] + T[k] for k in xrange(N/2)] + [even[k] - T[k] for k in xrange(N/2)]
    
y = dsp.SawtoothSignal().make_wave().ys[:2048]
fft = FFT(y)
ts = [i for i in xrange(len(fft))]
plt.plot(ts, np.abs(fft))
plt.xlabel("frequency")
plt.ylabel("amplitude")
plt.axis(xmax=2048)
plt.show()
```
